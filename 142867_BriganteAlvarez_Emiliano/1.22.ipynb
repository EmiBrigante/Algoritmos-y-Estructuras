```python
# Definición de la clase Libro
class Libro:
    def __init__(self, titulo, autor, anio_publicacion):
        self.titulo = titulo
        self.autor = autor
        self.anio_publicacion = anio_publicacion

    def mostrar_informacion(self):
        print(f"Título: {self.titulo}")
        print(f"Autor: {self.autor}")
        print(f"Año de Publicación: {self.anio_publicacion}")


# Definición de la clase Biblioteca
class Biblioteca:
    def __init__(self, nombre):
        self.nombre = nombre
        self.libros_disponibles = []

    def agregar_libro(self, libro):
        self.libros_disponibles.append(libro)
        print(f"{libro.titulo} ha sido agregado a la biblioteca.")

    def prestar_libro(self, titulo):
        for libro in self.libros_disponibles:
            if libro.titulo == titulo:
                self.libros_disponibles.remove(libro)
                print(f"{titulo} ha sido prestado.")
                return libro
        print(f"El libro {titulo} no está disponible.")

    def devolver_libro(self, libro):
        self.libros_disponibles.append(libro)
        print(f"{libro.titulo} ha sido devuelto a la biblioteca.")

    def mostrar_catalogo(self):
        print("Catálogo de libros disponibles:")
        for libro in self.libros_disponibles:
            libro.mostrar_informacion()
            print()


# Crear una instancia de la clase Biblioteca
mi_biblioteca = Biblioteca("Mi Biblioteca Favorita")

# Crear instancias de la clase Libro y agregarlas a mi_biblioteca
libro1 = Libro("El Principito", "Antoine de Saint-Exupéry", 1943)
libro2 = Libro("1984", "George Orwell", 1949)
libro3 = Libro("Don Quijote", "Miguel de Cervantes", 1605)

mi_biblioteca.agregar_libro(libro1)
mi_biblioteca.agregar_libro(libro2)
mi_biblioteca.agregar_libro(libro3)

# Mostrar el catálogo actualizado
mi_biblioteca.mostrar_catalogo()

# Realizar operaciones de préstamo y devolución
libro_prestado = mi_biblioteca.prestar_libro("1984")

# Mostrar el catálogo actualizado
mi_biblioteca.mostrar_catalogo()

# Devolver el libro prestado
mi_biblioteca.devolver_libro(libro_prestado)

# Mostrar el catálogo actualizado
mi_biblioteca.mostrar_catalogo()
```
